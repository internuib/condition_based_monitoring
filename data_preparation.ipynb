```python
from scipy.signal import butter, filtfilt
# Design a Butterworth low-pass filter
def butter_lowpass_filter(data, cutoff=500, fs=50000, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

# Apply filter to Underhang and Overhang signals
df_1sec["Underhang_Axial"] = butter_lowpass_filter(df_1sec["Underhang_Axial"])
df_1sec["Underhang_Radial"] = butter_lowpass_filter(df_1sec["Underhang_Radial"])
df_1sec["Underhang_Tangential"] = butter_lowpass_filter(df_1sec["Underhang_Tangential"])
df_1sec["Overhang_Axial"] = butter_lowpass_filter(df_1sec["Overhang_Axial"])
df_1sec["Overhang_Radial"] = butter_lowpass_filter(df_1sec["Overhang_Radial"])
df_1sec["Overhang_Tangential"] = butter_lowpass_filter(df_1sec["Overhang_Tangential"])

# Plot filtered data
plt.figure(figsize=(12, 6))

# Plot Underhang signals
plt.subplot(2, 1, 1)
plt.plot(time_1sec, df_1sec["Underhang_Axial"], label="Underhang Axial", color='r')
plt.plot(time_1sec, df_1sec["Underhang_Radial"], label="Underhang Radial", color='g')
plt.plot(time_1sec, df_1sec["Underhang_Tangential"], label="Underhang Tangential", color='m')
plt.xlabel("Time (s)")
plt.ylabel("Underhang Bearing")
plt.legend()
plt.grid()
plt.title("Filtered Underhang Bearing Signals (1 sec)")

# Plot Overhang signals
plt.subplot(2, 1, 2)
plt.plot(time_1sec, df_1sec["Overhang_Axial"], label="Overhang Axial", color='b')
plt.plot(time_1sec, df_1sec["Overhang_Radial"], label="Overhang Radial", color='c')
plt.plot(time_1sec, df_1sec["Overhang_Tangential"], label="Overhang Tangential", color='y')
plt.xlabel("Time (s)")
plt.ylabel("Overhang Bearing")
plt.legend()
plt.grid()
plt.title("Filtered Overhang Bearing Signals (1 sec)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
from scipy.fftpack import fft
from scipy.signal import welch

def process_vibration_data(csv_file, window_size=50000, overlap=5000, sampling_rate=50000):
    """
    Process a CSV file containing vibration data by applying sliding window segmentation
    and extracting time-domain and frequency-domain features.

    Parameters:
    - csv_file: Path to the CSV file
    - window_size: Number of samples per segment (default: 10,000)
    - overlap: Overlapping samples between windows (default: 5,000)
    - sampling_rate: Sampling rate of the signal (default: 50 kHz)

    Returns:
    - A DataFrame containing extracted features for each segment
    """

    # Load the CSV without headers
    df= pd.read_csv(csv_file, header=None)

    # Define column names
    df.columns = [
         "Tachometer", "Underhang_Axial", "Underhang_Radial", "Underhang_Tangential",
        "Overhang_Axial", "Overhang_Radial", "Overhang_Tangential", "Microphone"
    ]

    # Sensor columns for segmentation (excluding Tachometer & Microphone)
    sensor_columns = ['Underhang_Axial', 'Underhang_Radial', 'Underhang_Tangential',
                      'Overhang_Axial', 'Overhang_Radial', 'Overhang_Tangential']

    # Function to extract time-domain features
    def extract_time_features(signal):
        return {
            'Mean': np.mean(signal),
            'RMS': np.sqrt(np.mean(signal**2)),
            'Std': np.std(signal),
            'Kurtosis': kurtosis(signal),
            'Skewness': skew(signal),
            'Crest Factor': np.max(signal) / np.sqrt(np.mean(signal**2)),
            'Peak-to-Peak': np.ptp(signal)
        }

    # Function to extract frequency-domain features
    def extract_frequency_features(signal, sampling_rate):
        n = len(signal)
        freq_spectrum = np.abs(fft(signal))[:n // 2]
        freq_bins = np.linspace(0, sampling_rate / 2, len(freq_spectrum))

        freqs, psd = welch(signal, fs=sampling_rate)

        return {
            'FFT Energy': np.sum(freq_spectrum**2),
            'Spectral Entropy': -np.sum((psd / np.sum(psd)) * np.log2(psd / np.sum(psd))),
            'Dominant Frequency': freq_bins[np.argmax(freq_spectrum)],
            'Bandwidth': np.sqrt(np.sum((freq_bins - np.mean(freq_bins))**2 * freq_spectrum) / np.sum(freq_spectrum))
        }

    # Perform segmentation using sliding window
    features = []
    num_samples = len(df)

    for start in range(0, num_samples - window_size, overlap):
        end = start + window_size
        segment = df.iloc[start:end]

        # Extract features for each window
        window_features = {'Start_Index': start, 'End_Index': end}

        for col in sensor_columns:
            signal = segment[col].values
            time_features = extract_time_features(signal)
            freq_features = extract_frequency_features(signal, sampling_rate)

            # Prefix feature names with sensor name
            for key, value in {**time_features, **freq_features}.items():
                window_features[f"{col}_{key}"] = value

        features.append(window_features)

    # Convert to DataFrame
    segmented_df = pd.DataFrame(features)

    # Save segmented features to CSV
    output_file = csv_file.replace(".csv", "_features.csv")
    segmented_df.to_csv(output_file, index=False)

    print(f"Feature extraction completed. Saved as: {output_file}")
    
    return segmented_df

# Example Usage
df2 = process_vibration_data(r"D:\cbm_prj\normal\13.1072.csv")

```

```python
import pandas as pd
import glob

def combine_feature_csvs(folder_path, output_filename="normal_features.csv"):
    """
    Combine all CSV files in a folder into a single DataFrame and save as a new CSV.

    Parameters:
    - folder_path: Path where all the feature CSV files are located.
    - output_filename: Name of the output CSV file (default: "normal_features.csv")

    Returns:
    - A combined DataFrame containing all feature data.
    """

    # Find all CSV files in the specified folder
    csv_files = glob.glob(folder_path + "/*.csv")

    # List to store individual DataFrames
    dataframes = []

    # Load and append each CSV
    for file in csv_files:
        df = pd.read_csv(file)
        dataframes.append(df)

    # Combine all DataFrames
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    combined_df.to_csv(output_filename, index=False)

    print(f"Combined features saved as: {output_filename}")

    return combined_df

# Example Usage
combined_df = combine_feature_csvs(r"D:\cbm_prj\normal\normal_features")

```

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
from scipy.fftpack import fft
from scipy.signal import welch

def process_vibration_data(csv_file, window_size=25000, overlap=2500, sampling_rate=50000):
    """
    Process a CSV file containing vibration data by applying sliding window segmentation
    and extracting time-domain and frequency-domain features.

    Parameters:
    - csv_file: Path to the CSV file
    - window_size: Number of samples per segment (default: 10,000)
    - overlap: Overlapping samples between windows (default: 5,000)
    - sampling_rate: Sampling rate of the signal (default: 50 kHz)

    Returns:
    - A DataFrame containing extracted features for each segment
    """

    # Load the CSV without headers
    df = pd.read_csv(csv_file, header=None)

    # Define column names
    df.columns = [
        "Tachometer", "Underhang_Axial", "Underhang_Radial", "Underhang_Tangential",
        "Overhang_Axial", "Overhang_Radial", "Overhang_Tangential", "Microphone"
    ]

    # Sensor columns for segmentation (excluding Tachometer & Microphone)
    sensor_columns = ['Underhang_Axial', 'Underhang_Radial', 'Underhang_Tangential',
                      'Overhang_Axial', 'Overhang_Radial', 'Overhang_Tangential']

    # Function to extract time-domain features
    def extract_time_features(signal):
        return {
            'Mean': np.mean(signal),
            'RMS': np.sqrt(np.mean(signal**2)),
            'Std': np.std(signal),
            'Kurtosis': kurtosis(signal),
            'Skewness': skew(signal),
            'Crest Factor': np.max(signal) / np.sqrt(np.mean(signal**2)),
            'Peak-to-Peak': np.ptp(signal)
        }

    # Function to extract frequency-domain features
    def extract_frequency_features(signal, sampling_rate):
        n = len(signal)
        freq_spectrum = np.abs(fft(signal))[:n // 2]
        freq_bins = np.linspace(0, sampling_rate / 2, len(freq_spectrum))

        freqs, psd = welch(signal, fs=sampling_rate)

        return {
            'FFT Energy': np.sum(freq_spectrum**2),
            'Spectral Entropy': -np.sum((psd / np.sum(psd)) * np.log2(psd / np.sum(psd))),
            'Dominant Frequency': freq_bins[np.argmax(freq_spectrum)],
            'Bandwidth': np.sqrt(np.sum((freq_bins - np.mean(freq_bins))**2 * freq_spectrum) / np.sum(freq_spectrum))
        }

    # Perform segmentation using sliding window
    features = []
    num_samples = len(df)

    for start in range(0, num_samples - window_size, overlap):
        end = start + window_size
        segment = df.iloc[start:end]

        # Extract features for each window
        window_features = {'Start_Index': start, 'End_Index': end, 'File': os.path.basename(csv_file)}

        for col in sensor_columns:
            signal = segment[col].values
            time_features = extract_time_features(signal)
            freq_features = extract_frequency_features(signal, sampling_rate)

            # Prefix feature names with sensor name
            for key, value in {**time_features, **freq_features}.items():
                window_features[f"{col}_{key}"] = value

        features.append(window_features)

    # Convert to DataFrame
    segmented_df = pd.DataFrame(features)

    return segmented_df

def process_all_files(folder_path, output_file):
    """
    Process all CSV files in a folder, extract features, and save to a single CSV file.

    Parameters:
    - folder_path: Path to the folder containing CSV files
    - output_file: Name of the final combined CSV file
    """

    all_features = []

    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            file_path = os.path.join(folder_path, file)
            print(f"Processing: {file_path}")
            segmented_df = process_vibration_data(file_path)
            all_features.append(segmented_df)

    # Combine all extracted features into a single DataFrame
    combined_features = pd.concat(all_features, ignore_index=True)

    # Save to CSV
    combined_features.to_csv(output_file, index=False)
    print(f"Feature extraction completed. Saved to {output_file}")

# Folder containing CSV files
folder_path = r"D:\cbm_prj\horizontal-misalignment\0.5mm"  # Change this to your actual folder path
output_file = "0.5mm_horizontal_features.csv"

# Process all CSV files in the folder
process_all_files(folder_path, output_file)
```

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
from scipy.fftpack import fft
from scipy.signal import welch

def process_vibration_data(csv_file, window_size=25000, overlap=2500, sampling_rate=50000):
    """
    Process a CSV file containing vibration data by applying sliding window segmentation
    and extracting time-domain and frequency-domain features.

    Parameters:
    - csv_file: Path to the CSV file
    - window_size: Number of samples per segment (default: 10,000)
    - overlap: Overlapping samples between windows (default: 5,000)
    - sampling_rate: Sampling rate of the signal (default: 50 kHz)

    Returns:
    - A DataFrame containing extracted features for each segment
    """

    # Load the CSV without headers
    df = pd.read_csv(csv_file, header=None)

    # Define column names
    df.columns = [
        "Tachometer", "Underhang_Axial", "Underhang_Radial", "Underhang_Tangential",
        "Overhang_Axial", "Overhang_Radial", "Overhang_Tangential", "Microphone"
    ]

    # Sensor columns for segmentation (excluding Tachometer & Microphone)
    sensor_columns = ['Underhang_Axial', 'Underhang_Radial', 'Underhang_Tangential',
                      'Overhang_Axial', 'Overhang_Radial', 'Overhang_Tangential']

    # Function to extract time-domain features
    def extract_time_features(signal):
        return {
            'Mean': np.mean(signal),
            'RMS': np.sqrt(np.mean(signal**2)),
            'Std': np.std(signal),
            'Kurtosis': kurtosis(signal),
            'Skewness': skew(signal),
            'Crest Factor': np.max(signal) / np.sqrt(np.mean(signal**2)),
            'Peak-to-Peak': np.ptp(signal)
        }

    # Function to extract frequency-domain features
    def extract_frequency_features(signal, sampling_rate):
        n = len(signal)
        freq_spectrum = np.abs(fft(signal))[:n // 2]
        freq_bins = np.linspace(0, sampling_rate / 2, len(freq_spectrum))

        freqs, psd = welch(signal, fs=sampling_rate)

        return {
            'FFT Energy': np.sum(freq_spectrum**2),
            'Spectral Entropy': -np.sum((psd / np.sum(psd)) * np.log2(psd / np.sum(psd))),
            'Dominant Frequency': freq_bins[np.argmax(freq_spectrum)],
            'Bandwidth': np.sqrt(np.sum((freq_bins - np.mean(freq_bins))**2 * freq_spectrum) / np.sum(freq_spectrum))
        }

    # Perform segmentation using sliding window
    features = []
    num_samples = len(df)

    for start in range(0, num_samples - window_size, overlap):
        end = start + window_size
        segment = df.iloc[start:end]

        # Extract features for each window
        window_features = {'Start_Index': start, 'End_Index': end, 'File': os.path.basename(csv_file)}

        for col in sensor_columns:
            signal = segment[col].values
            time_features = extract_time_features(signal)
            freq_features = extract_frequency_features(signal, sampling_rate)

            # Prefix feature names with sensor name
            for key, value in {**time_features, **freq_features}.items():
                window_features[f"{col}_{key}"] = value

        features.append(window_features)

    # Convert to DataFrame
    segmented_df = pd.DataFrame(features)

    return segmented_df

def process_all_files(folder_path, output_file):
    """
    Process all CSV files in a folder, extract features, and save to a single CSV file.

    Parameters:
    - folder_path: Path to the folder containing CSV files
    - output_file: Name of the final combined CSV file
    """

    all_features = []

    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            file_path = os.path.join(folder_path, file)
            print(f"Processing: {file_path}")
            segmented_df = process_vibration_data(file_path)
            all_features.append(segmented_df)

    # Combine all extracted features into a single DataFrame
    combined_features = pd.concat(all_features, ignore_index=True)

    # Save to CSV
    combined_features.to_csv(output_file, index=False)
    print(f"Feature extraction completed. Saved to {output_file}")

# Folder containing CSV files
folder_path = r"D:\cbm_prj\horizontal-misalignment\0.5mm"  # Change this to your actual folder path
output_file = "0.5mm_horizontal_features.csv"

# Process all CSV files in the folder
process_all_files(folder_path, output_file)
```

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew

# Folder containing normal data
folder_path = r"D:\cbm_prj\normal"

# Sampling rate and window parameters
sampling_rate = 50000  # 50 kHz
window_size = int(0.1 * sampling_rate)  # 0.1 sec = 5000 samples
overlap = int(0.5 * window_size)  # 50% overlap = 2500 samples

# Sensor columns (excluding Tachometer & Microphone)
sensor_columns = [
    "Underhang_Axial", "Underhang_Radial", "Underhang_Tangential",
    "Overhang_Axial", "Overhang_Radial", "Overhang_Tangential"
]

# Initialize a list to store feature data
feature_data = []

# Process each CSV file
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)

        # Load CSV file
        df = pd.read_csv(file_path)

        # Rename columns (if not already named)
        df.columns = ["Tachometer", "Underhang_Axial", "Underhang_Radial", "Underhang_Tangential",
                      "Overhang_Axial", "Overhang_Radial", "Overhang_Tangential", "Microphone"]

        # Extract features using sliding windows
        for start in range(0, len(df) - window_size, overlap):
            window = df.iloc[start: start + window_size]

            features = {}
            features['filename'] = file
            features['window_start'] = start
            features['window_end'] = start + window_size

            for col in sensor_columns:
                signal = window[col]
                features[f"{col}_mean"] = np.mean(signal)
                features[f"{col}_std"] = np.std(signal)
                features[f"{col}_rms"] = np.sqrt(np.mean(signal**2))
                features[f"{col}_kurtosis"] = kurtosis(signal)
                features[f"{col}_skewness"] = skew(signal)
                features[f"{col}_peak_to_peak"] = np.ptp(signal)  # Peak-to-peak amplitude

            # Append features to list
            feature_data.append(features)

# Convert to DataFrame
features_df = pd.DataFrame(feature_data)

# Save to CSV (Optional)
features_df.to_csv("normal_windowed_features.csv", index=False)


```
